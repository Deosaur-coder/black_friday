```python
import pandas as pd 
import seaborn as sns
import matplotlib as plt
import numpy as np
```

```python
#checking both test and training data and importing it 
df_test=pd.read_csv("test.csv")

df_test.head()
#df_test.info()
```

```python
df_train=pd.read_csv("train.csv")

df_train.head()
```

```python
#merging both train data and test data and removing/deleting  irrelavent coloums
#1st method
df=df_train.append(df_test)
df.info()

```

```python
#deleting user id cuz it has nothing to do with alanaysis and relations bw coloumns 
df.drop(['User_ID'],axis=1,inplace=True)
df.head()
```

```python
df_gender=df['Gender']
df_gender.head()
#we have only 2 cat male and female ,so we can use getdummies using pandas or map using python
```

```python
df["Gender"]=pd.get_dummies(df["Gender"],drop_first=1)
df.head()
```

```python
df.head()
```

```python
df_age=df["Age"]
df_age.unique()
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])

df['Age'].unique()
df.head(30)
```

```python
#this has 3 feautures so it needs 2 clumns to print so save it in a seprate df and concat it aT THE END

df_city=pd.get_dummies(df["City_Category"],drop_first=1)
df_city.head()
```

```python
# 2 coloumns will be added in the last part )
pd.concat([df,df_city],axis=1)
```
